```python
#lets map the features with meaning full names.
#classes: edible=e, poisonous=p)
df['class'] = df['class'].replace({'e':'Edible','p':'Poisonous'})
```

```python
#cap-shape: bell=b,conical=c,convex=x,flat=f, knobbed=k,sunken=s
df['cap-shape'] = df['cap-shape'].replace({'b':'bell',
                                           'c':'conical',
                                           'x':'convex',
                                           'f':'flat',
                                           'k':'knobbed',
                                           's':'sunken'})
```

```python
#cap-surface: fibrous=f,grooves=g,scaly=y,smooth=s
df['cap-surface'] = df['cap-surface'].replace({
                                           'y':'scaly',
                                           'g':'grooves',
                                           'f':'fibrous',
                                           's':'smooth'})
```

```python
#cap-color: brown=n,buff=b,cinnamon=c,gray=g,green=r,pink=p,purple=u,red=e,white=w,yellow=y
df['cap-color'] = df['cap-color'].replace({'n':'brown','b':'buff',
                                           'c':'cinnamon','u':'purple',
                                           'e':'red','w':'white',
                                           'y':'yellow','g':'gray',
                                           'r':'green','p':'pink'})
```

```python
#bruises: bruises=t,no=f
df['bruises'] = df['bruises'].replace({'t':'bruises','f':'no'})                                          
```

```python
#odor: almond=a,anise=l,creosote=c,fishy=y,foul=f,musty=m,none=n,pungent=p,spicy=s
df['odor'] = df['odor'].replace({'a':'almond','l':'anise',
                                 'c':'creosote','y':'fishy',
                                 'f':'foul','m':'musty',
                                 'n':'none','p':'pungent',
                                 's':'spicy'})
```

```python
# gill-attachment: attached=a,descending=d,free=f,notched=n
df['gill-attachment'] = df['gill-attachment'].replace({'n':'notched','f':'free',
                                                       'd':'descending','a':'attached'})
```

```python
#gill-spacing: close=c,crowded=w,distant=d
df['gill-spacing'] = df['gill-spacing'].replace({'c':'close','w':'crowded',
                                                 'd':'distant'})
```

```python
#gill-size: broad=b,narrow=n
df['gill-size'] = df['gill-size'].replace({'n':'narrow','b':'broad'})
```

```python
#gill-color: black=k,brown=n,buff=b,chocolate=h,gray=g, green=r,orange=o,pink=p,purple=u,red=e,white=w,yellow=y
df['gill-color'] = df['gill-color'].replace({'k':'black','n':'brown',
                                             'b':'buff','h':'chocolate',
                                             'g':'gray','r':'green',
                                             'o':'orange','p':'pink',
                                             'u':'purple','e':'red',
                                             'w':'white','y':'yellow'})
```

```python
#stalk-shape: enlarging=e,tapering=t
df['stalk-shape'] = df['stalk-shape'].replace({'e':'enlarging','t':'tapering'})
```

```python
#stalk-root: bulbous=b,club=c,cup=u,equal=e,rhizomorphs=z,rooted=r,missing=?
df['stalk-root'] = df['stalk-root'].replace({'b':'bulbous','c':'club',
                                             'u':'cup','e':'equal',
                                             'z':'rhizomorphs','r':'rooted',
                                             '?':np.nan})
```

```python
#stalk-surface-above-ring: fibrous=f,scaly=y,silky=k,smooth=s
df['stalk-surface-above-ring'] = df['stalk-surface-above-ring'].replace({'f':'fibrous','y':'scaly',
                                                                         'k':'silky','s':'smooth'})
```

```python
#stalk-surface-below-ring: fibrous=f,scaly=y,silky=k,smooth=s
df['stalk-surface-below-ring'] = df['stalk-surface-below-ring'].replace({'f':'fibrous','y':'scaly',
                                                                         'k':'silky','s':'smooth'})
```

```python
#stalk-color-above-ring: brown=n,buff=b,cinnamon=c,gray=g,orange=o,pink=p,red=e,white=w,yellow=y
df['stalk-color-above-ring'] = df['stalk-color-above-ring'].replace({'n':'brown','b':'buff',
                                                                     'c':'cinnamon','e':'red',
                                                                     'w':'white','p':'pink',
                                                                     'y':'yellow','g':'gray',
                                                                     'o':'orange'})
```

```python
#stalk-color-below-ring: brown=n,buff=b,cinnamon=c,gray=g,orange=o,pink=p,red=e,white=w,yellow=y
df['stalk-color-below-ring'] = df['stalk-color-below-ring'].replace({'n':'brown','b':'buff',
                                                                     'c':'cinnamon','e':'red',
                                                                     'w':'white','p':'pink',
                                                                     'y':'yellow','g':'gray',
                                                                     'o':'orange'})
```

```python
#veil-type: partial=p,universal=u
df['veil-type'] = df['veil-type'].replace({'p':'partial','u':'universal'})
```

```python
#veil-color: brown=n,orange=o,white=w,yellow=y
df['veil-color'] = df['veil-color'].replace({'n':'brown',                                           
                                             'w':'white',
                                             'y':'yellow',
                                             'o':'orange'})
```

```python
#ring-number: none=n,one=o,two=t
df['ring-number'] = df['ring-number'].replace({'n':'none','o':'one','t':'two'})
```

```python
#ring-type: cobwebby=c,evanescent=e,flaring=f,large=l,none=n,pendant=p,sheathing=s,zone=z
df['ring-type'] = df['ring-type'].replace({'c':'cobwebby','e':'evanescent',
                                           'f':'flaring','l':'large',
                                           'n':'none','p':'pendant',
                                           's':'sheathing','z':'zone'})
```

```python
#spore-print-color: black=k,brown=n,buff=b,chocolate=h,green=r,orange=o,purple=u,white=w,yellow=y
df['spore-print-color'] = df['spore-print-color'].replace({'n':'brown','b':'buff',
                                                           'k':'black','u':'purple',
                                                           'o':'orange','w':'white',
                                                           'y':'yellow','h':'chocolate',
                                                           'r':'green'})
```

```python
#population: abundant=a,clustered=c,numerous=n,scattered=s,several=v,solitary=y
df['population'] = df['population'].replace({'a':'abundant','c':'clustered',
                                             'n':'numerous','s':'scattered',
                                             'v':'several','y':'solitary'})
```

```python
#habitat: grasses=g,leaves=l,meadows=m,paths=p,urban=u,waste=w,woods=d
df['habitat'] = df['habitat'].replace({'g':'grasses','l':'leaves',
                                           'm':'meadows','p':'paths',
                                           'u':'urban','w':'waste',
                                           'd':'woods'})
```

```python
%matplotlib inline
import category_encoders as ce
from sklearn.impute import SimpleImputer
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn import metrics
```

```python
target = 'class'
# Split train,test.
train, test = train_test_split(df, train_size=0.80, test_size=0.20, 
                              stratify=df[target], random_state=42)
train.shape , test.shape
```

```python
#Spliting train into train ,val.
train, val = train_test_split(train, train_size=0.80, test_size=0.20, 
                              stratify=train[target], random_state=42)
train.shape , val.shape
```

```python
# Get a dataframe with all train columns except the target
train_features =train.drop(columns=[target])
val_features =  val.drop(columns=[target])
test_features = test.drop(columns=[target])

train_features.shape,val_features.shape,test_features.shape
```

```python
#used_features=train.drop(columns=['class','odor','gill-size','ring-type','gill-attachment','gill-color','veil-color','veil-type'])
features= ['bruises','odor','population','habitat']
features
```

```python
# Arrange data into X features matrix and y target vector 
X_train = train_features[features]
y_train = train[target]
X_val = val_features[features]
y_val = val[target]
X_test = test_features[features]
y_test = test[target]
```

```python
print(X_train.sample(5))
print(X_val.sample(5))
```

```python
processor = make_pipeline(
    ce.OrdinalEncoder(), 
    SimpleImputer(strategy='median')
)

X_train_processed = processor.fit_transform(X_train)
X_val_processed = processor.transform(X_val)
X_test_processed = processor.transform(X_test)

model_forest = RandomForestClassifier(random_state=42, n_jobs=1)
```

```python
# Fit on train, score on val
model_forest.fit(X_train_processed, y_train)
print('Validation Accuracy', model_forest.score(X_val_processed, y_val))
```

```python
# Predict on test
y_pred = model_forest.predict(X_test_processed)
y_pred
```

```python
#testing single random values to get prediction on random new data
new_input = [['bruises','fishy','scattered','leaves']]

new_data = pd.DataFrame(new_input, columns=['bruises','odor','population','habitat'])
new_data
```

```python
new_data_processed = processor.transform(new_data)
```

```python
model_forest.predict(new_data_processed)
```

```python
#Acuracy on test data
print("Random Forest Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
#confusion metrics 
cf_metrics = metrics.confusion_matrix(y_test,y_pred)
cf_metrics
```

```python
#ploting confusion metrics with custom labels that contains required information
group_names = ['True Neg','False Pos','False Neg','True Pos']
group_counts = ['{0:0.0f}'.format(value) for value in
                cf_metrics.flatten()]
group_percentages = ['{0:.2%}'.format(value) for value in
                     cf_metrics.flatten()/np.sum(cf_metrics)]
labels = [f"{v1}\n{v2}\n{v3}" for v1, v2, v3 in
          zip(group_names,group_counts,group_percentages)]
labels = np.asarray(labels).reshape(2,2)
catagories = ['Edible', 'Poisonous']
 #location = {'center', 'left', 'right'},str,optional
   
ax = plt.axes()


cm2 = sns.heatmap(cf_metrics, annot=labels, fmt='', cmap='Blues',ax= ax);
ax.set_title('Confusion Matrix of RandomForest',fontdict={'fontsize':18 ,  
             'color' : 'red',  
             'verticalalignment': 'baseline',  
             'horizontalalignment':'center'})
ax.set_xlabel('Predicted Data', fontsize = 15)
ax.set_ylabel('True Data', fontsize = 15)
sticklabel_x =['Edible','Poisonious']
sticklabel_y =['Edible','Poisonious']
ax.set_xticklabels(sticklabel_x)
ax.set_yticklabels(sticklabel_y)
plt.savefig('cm2.png',dpi=300)
plt.show()
```

```python
# Create feature matrix and target vector
# X_train_processed
# y_train
# X_val_processed
# y_val
# X_test_processed
# y_test
from sklearn.model_selection import validation_curve
# Create range of values for parameter
param_range = np.arange(1, 20, 2)

# Calculate accuracy on training and test set using range of parameter values
train_scores, test_scores = validation_curve(model_forest, 
                                             X_train_processed, 
                                             y_train, 
                                             param_name="n_estimators", 
                                             param_range=param_range,
                                             cv=3, 
                                             scoring="accuracy", 
                                             n_jobs=-1)


# Calculate mean and standard deviation for training set scores
train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)

# Calculate mean and standard deviation for test set scores
test_mean = np.mean(test_scores, axis=1)
test_std = np.std(test_scores, axis=1)

# Plot mean accuracy scores for training and test sets
plt.plot(param_range, train_mean, label="Training score", color="black")
plt.plot(param_range, test_mean, label="Cross-validation score", color="dimgrey")

# Plot accurancy bands for training and test sets
plt.fill_between(param_range, train_mean - train_std, train_mean + train_std, color="gray")
plt.fill_between(param_range, test_mean - test_std, test_mean + test_std, color="gainsboro")

# Create plot
plt.title("Validation Curve With Random Forest")
plt.xlabel("Number Of Trees")
plt.ylabel("Accuracy Score")
plt.tight_layout()
plt.legend(loc="best")
plt.savefig('vc.jpeg',dpi=200,bbox_inches='tight')
plt.show()
```
